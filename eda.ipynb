```python
plt.figure(figsize=(10,9))
hs = sns.countplot(df['Are you a current high school or university student?'], palette=['green', 'red'])
total = len(df['Are you a current high school or university student?'])
for p in hs.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    # hs.annotate(percentage, (x, y),ha='center')
    hs.annotate(percentage,  (x, y),
                   ha = 'center', va = 'center', 
                   size=10,
                   xytext = (-110, 10), 
                   textcoords = 'offset points')
```
